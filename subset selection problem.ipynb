```python
from itertools import combinations

# Given set of numbers
nums = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Generate all subsets of size 5 whose sum is zero
subset_size = 5
valid_subsets = []

for subset in combinations(nums, subset_size):
    if sum(subset) == 0:
        valid_subsets.append(subset)

# Print all valid subsets     #vs=valid sets
for vs in valid_subsets:
    print(vs)
```
